```python
import matplotlib.pyplot as plt

# Count occurrences of each gender
gender_counts = df['Gender'].value_counts()

# Plot the pie chart
plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightcoral'],explode =[0,.02])
plt.title('Gender Distribution')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter dataset for employees who left
df_filtered = df[df['Attrition'] == 'Yes']

# Set figure size
plt.figure(figsize=(12,6))

# Create grouped bar chart
sns.countplot(data=df_filtered, x='EducationField', hue='Gender', palette='Set2')

# Labels and title
plt.xlabel("Education Field")
plt.ylabel("Number of Employees who left")
plt.title("Attrition by Education Field and Gender")
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.legend(title="Gender")

plt.show()

```

```python
import matplotlib.pyplot as plt

# Group by Overtime and Attrition and count occurrences
overtime_attrition_counts = df.groupby(["OverTime", "Attrition"]).size()

# Convert to DataFrame for better visualization
overtime_attrition_counts = overtime_attrition_counts.reset_index(name="Count")

# Create labels
labels = [
    f"Overtime - Yes, Attrition - Yes",
    f"Overtime - Yes, Attrition - No",
    f"Overtime - No, Attrition - Yes",
    f"Overtime - No, Attrition - No"
]

# Create Pie Chart
plt.figure(figsize=(8,8))
plt.pie(
    overtime_attrition_counts["Count"], 
    labels=labels, 
    autopct='%1.1f%%', 
    startangle=90, 
    colors=['lightcoral', 'skyblue', 'lightgreen', 'gold'], 
    explode=[0.05, 0.05, 0.05, 0.05]
)

plt.title("Overtime & Attrition Distribution")
plt.show()
```

```python
# Group by MaritalStatus and Attrition
attrition_status = df.groupby(['MaritalStatus', 'Attrition']).size().unstack()

# Plot
attrition_status.plot(kind='bar', stacked=True, figsize=(8,5), colormap='coolwarm')

# Labels and title
plt.xlabel("Marital Status")
plt.ylabel("Count of Employees")
plt.title("Attrition Breakdown by Marital Status")
plt.legend(title="Attrition")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Get counts for each category
attrition_counts = df.groupby(['BusinessTravel', 'Attrition']).size().unstack()

# Plot stacked bar chart
ax = attrition_counts.plot(kind='bar', stacked=True, figsize=(10, 6), colormap="viridis")

# Annotate bars with values
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type="center", fontsize=10, color="white", weight="bold")

# Labels and title
plt.title("Attrition by Business Travel Type")
plt.xlabel("Business Travel Type")
plt.ylabel("Number of Employees")
plt.legend(title="Attrition", labels=["No", "Yes"])
plt.xticks(rotation=45)

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=df, x="TotalWorkingYears", y="MonthlyIncome", 
    hue="Attrition", size="EnvironmentSatisfaction",
    alpha=0.6, sizes=(50, 500), palette="coolwarm"
)

# Titles and labels
plt.title("Employee Attrition: Age vs Monthly Income (Bubble Chart)")
plt.xlabel("Age")
plt.ylabel("Monthly Income")
plt.legend(title="Attrition")
plt.show()
```
